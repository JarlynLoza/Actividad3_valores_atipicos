```python
#Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_2021 =pd.read_excel ('Gastos y costos 20-23.xlsx', sheet_name= '2021', skiprows= 3)
df_2021
```

```python
df_2021.info()
```

```python
#Corroboramos valores nulos
valores_nulos=df_2021.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con bfill
data21=df_2021.fillna(method="bfill")
data21
```

```python
#Corroboramos valores nulos
valores_nulos=data21.isnull().sum()
valores_nulos
```

```python
# Sustituir valores nulos por valores no nulos hacia adelante "forward fill" (ffill)
#Filtro por columnas
data21["POLIZA"]=data21["POLIZA"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data21.isnull().sum()
valores_nulos
```

```python
#filtro por columnas (cualitativas y cuantitativas)
cuanti=data21.iloc[:,8:14]
cuali=data21.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
```

```python
#Diagrama de caja o bigote de cada columna del dataframe
fig=plt.figure(figsize=(15,8))
cuanti.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #Dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=cuanti

percentile25=y.quantile(0.25) #Q!
percentile75=y.quantile(0.75) #Q3
iqr= percentile75- percentile25

Limite_Superior_iqr=percentile75+1.5*iqr
Limite_Inferior_iqr=percentile25-1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos los outliers
##Convertir outliers en nulos
outliers_iqr= cuanti[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del dataframe
cuanti_clean_iqr=cuanti[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
cuanti_clean_iqr
```

```python
#Corroborar valores nulos
valores_nulos=cuanti_clean_iqr.isnull().sum()
valores_nulos
```

```python
#Se Reemplazan los valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
cuanti_clean_iqr1=cuanti.copy()
cuanti_clean_iqr1=cuanti_clean_iqr.fillna(round(cuanti.mean(),1))
cuanti_clean_iqr1
```

```python
#Corroborar valores nulos
valores_nulos=cuanti_clean_iqr1.isnull().sum()
valores_nulos
```

```python
#Concatenar columnas cualitativas con columnas cuantitativas
#cuali= data20['cuali']
Datos_limpios_2021= pd.concat([cuanti,cuali], axis=1)
Datos_limpios_2021
```

```python
#Convertir DataFrame a csv
Datos_limpios_2021.to_csv("Frame_limpio_2021.csv")
```

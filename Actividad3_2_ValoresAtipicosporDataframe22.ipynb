```python
#Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga de un archivo .xlsx desde un libro con varias páginas
df_2022=pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name='2022')
df_2022
```

```python
df_2022 =pd.read_excel ('Gastos y costos 20-23.xlsx', sheet_name= '2022', skiprows= 3)
df_2022
```

```python
#Eliminar columnas innecesarias
#axis= 1 para columnas
#axis= 0 para filas
df_2022= df_2022.drop(['Otros'], axis=1)
df_2022
```

```python
df_2022.info()
```

```python
#Corroboramos valores nulos
valores_nulos=df_2022.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores nulos del dataframe con bfill
data22=df_2022.fillna(method="bfill")
```

```python
#Corroboramos valores nulos
valores_nulos=data22.isnull().sum()
valores_nulos
```

```python
#Eliminar datos nulos del resto de las columnas
data22["Poliza"]=data22["Poliza"].fillna(method="ffill")
data22["MP"]=data22["MP"].fillna(method="ffill")
data22["TC"]=data22["TC"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos=data22.isnull().sum()
valores_nulos
```

```python
data22.info()
```

```python
#filtro por columnas (cualitativas y cuantitativas)
cuanti=data22.iloc[:,8:15]
cuali=data22.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
```

```python
#Diagrama de caja o bigote de cada columna del dataframe
fig=plt.figure(figsize=(15,8))
cuanti.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #Dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=cuanti
Limite_Superior=y.mean()+ 3*y.std()
Limite_Inferior=y.mean()-3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos los outliers dentro del dataframe
outliers= cuanti[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el dataframe
df22=cuanti[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df22
```

```python
#Corroboramos valores nulos
valores_nulos=cuanti.isnull().sum()
valores_nulos
```

```python
#Concatenar columnas cualitativas con columnas cuantitativas
#cuali= data20['cuali']
Datos_limpios_2022= pd.concat([cuanti,cuali], axis=1)
Datos_limpios_2022
```

```python
#Convertir DataFrame a csv
Datos_limpios_2022.to_csv("Frame_limpio_2022.csv")
```

```python
#Importamos librer√≠as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cargar desde un archivo .csv sin indice
data=pd.read_csv('VentasTotalesLimpio.csv')
data
```

```python
#Se corrobora valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Eliminar primera fila o filas innecesarias
data1= data.drop(['Unnamed: 0'],axis=1)
```

```python
#Se hace una copia del DataFrame
data2=data1.copy()
data2
```

##Columna ventas_precios_corrientes

```python
#Realizar diagrama de caja bigote
fig1= plt.figure(figsize=(10,6))
plt.boxplot(data2["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()#Dibujamos el diagrama
```

```python
#M√©todo aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data2["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q!
percentile75=y.quantile(0.75) #Q3
iqr= percentile75- percentile25

Limite_Superior_iqr=percentile75+1.5*iqr
Limite_Inferior_iqr=percentile25-1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos los outliers
outliers_iqr= data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr

```

```python
#Obtenemos datos limpios
col1=data1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
col1
```

```python
#Realizar diagrama de caja bigote
fig1= plt.figure(figsize=(10,6))
plt.boxplot(col1["ventas_precios_corrientes"])
plt.title("Outliers de ventas_precios_corrientes")
plt.show()#Dibujamos el diagrama
```

```python
#Convertir DataFrame a csv
col1.to_csv("ventas_precios_corrientes.csv")
```

```python
from google.colab import files

files.download("ventas_precios_corrientes.csv")
```

##ventas_precios_constantes

```python
#Realizar diagrama de caja bigote
fig= plt.figure(figsize=(10,6))
plt.boxplot(data2["ventas_precios_constantes"])
plt.title("Outliers de ventas_precios_constantes")
plt.show()#Dibujamos el diagrama
```

```python
#M√©todo aplicando desviaci√≥n est√°ndar. Encuentro los valores extremos
y=data2["ventas_precios_constantes"]
Limite_Superior=y.mean()+ 3*y.std()
Limite_Inferior=y.mean()-3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos los outliers
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Realizamos diagrama de caja bigote
fig= plt.figure(figsize=(5,3))
plt.boxplot(col2["ventas_precios_constantes"])
plt.title("Outliers de ventas_precios_constantes")
plt.show()#Dibujamos el diagrama
```
